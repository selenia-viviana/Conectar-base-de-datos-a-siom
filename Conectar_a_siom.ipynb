```python
# conectarse a base de datos
cnx = mysql.connector.connect(
        user="root",
        password="viviana",
        host="localhost",
        database="telefonica_siom",
        connection_timeout=10000)

# query de interes

query = '''SELECT 
os.orse_id, os.orse_fecha_creacion, fit.foit_opciones, e.empl_nombre, e.empl_direccion, e.empl_nemonico
FROM formulario_valor fv
left join formulario_item fit on fv.foit_id = fit.foit_id 
left join rel_informe_formulario_respuesta rifr on fv.fore_id = rifr.fore_id
left join formulario_respuesta fr on fv.fore_id
left join rel_orden_servicio_formulario rosf on rosf.form_id = fr.form_id 
left join orden_servicio os on os.orse_id = rosf.orse_id 
left join emplazamiento e on e.empl_id = os.empl_id 
where fv.foit_id = 21062'''


# se hace un dataframe
db_cursor = cnx.cursor()
db_cursor.execute(query)
rows =db_cursor.fetchall()
df = pd.DataFrame(rows, columns = db_cursor.column_names)
df.columns
```

```python
def extract_labels(json_str):
    try:
        data = json.loads(json_str)
        return data['labels']
    except (json.JSONDecodeError, KeyError):
        return [] 
```
